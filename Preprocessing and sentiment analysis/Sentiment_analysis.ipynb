```
import pandas as pd

id=pd.read_csv('/content/id.csv')
ids=id['id']
ls=pd.read_csv('reviews_dec18.csv')
#result = pd.concat([ls,ids ], axis=1, join='inner',on='id')
ls['del']=ls['listing_id'].isin(ids)
news=ls.loc[ls['del']==True]
news.dropna(inplace=True)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(news,news['id'],
                                                     test_size=0.10,
                                                     )
ls=X_test
```

```
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```

sid.polarity_scores("this room is clean")
```

```
import pandas as pd
import re
ls= news
ls.dropna(inplace=True)
print(ls.isna().sum())
def fun(row):
    res=fun.var.polarity_scores(row)
    return res['compound']
fun.var= sid
ls['res']=ls.comments.apply(fun)
ls.head()
```

```
x=ls.groupby('listing_id',as_index=False).mean()
x.to_csv('final.csv')
```
